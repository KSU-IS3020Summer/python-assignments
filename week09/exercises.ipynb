### Exercise 2: BankAccount Class
Create a `BankAccount` class with:
- `__init__(self, owner, balance=0)`
- `deposit(self, amount)`: add if positive, return new balance
- `withdraw(self, amount)`: subtract if sufficient, return new balance. If not enough, return `"Insufficient funds"`
- `transfer(self, other_account, amount)`: withdraw from self, deposit to other. Return True/False.
- `__str__(self)`: return `"Account(owner: $balance)"`

```python
# Exercise 2
class BankAccount:
    pass

# Test
acc1 = BankAccount("Alice", 1000)
acc2 = BankAccount("Bob", 500)
print(acc1)                    # Account(Alice: $1000)
print(acc1.deposit(200))       # 1200
print(acc1.withdraw(100))      # 1100
print(acc1.withdraw(5000))     # Insufficient funds
print(acc1.transfer(acc2, 300))# True
print(acc1)                    # Account(Alice: $800)
print(acc2)                    # Account(Bob: $800)
```

### Exercise 3: TodoList Class
Create a `TodoList` class with:
- `__init__(self)`: empty task list
- `add_task(self, task)`: add `{"task": task, "done": False}`
- `complete_task(self, index)`: mark as done. Return False if invalid index.
- `get_pending(self)`: return list of undone task strings
- `get_completed(self)`: return list of done task strings
- `__str__(self)`: return `"TodoList: X pending, Y completed"`

```python
# Exercise 3
class TodoList:
    pass

# Test
todo = TodoList()
todo.add_task("Buy groceries")
todo.add_task("Clean room")
todo.add_task("Study Python")
todo.complete_task(0)
print(todo)                   # TodoList: 2 pending, 1 completed
print(todo.get_pending())     # ["Clean room", "Study Python"]
print(todo.get_completed())   # ["Buy groceries"]
```

- `get_value(self)`: return total inventory value
- `low_stock(self, threshold=5)`: return list of item names with quantity <= threshold

```python
# Exercise 4
class Inventory:
    pass

# Test
shop = Inventory()
shop.add_item("Laptop", 999.99, 10)
shop.add_item("Mouse", 29.99, 50)
shop.add_item("Keyboard", 79.99, 3)
print(shop.sell_item("Mouse", 5))     # 149.95
print(shop.get_value())                # total value
print(shop.low_stock())                # ["Keyboard"]
```
